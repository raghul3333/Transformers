```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

features = df[['Engine rpm', 'Lub oil pressure', 'Fuel pressure', 'Coolant pressure', 'lub oil temp', 'Coolant temp']]
target = df[['Engine Condition']]

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.15, random_state=42)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## ANN (Creating Embeddings)

```python
import torch
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import torch.nn as nn
from tqdm import tqdm
from imblearn.under_sampling import RandomUnderSampler
from sklearn.utils import shuffle

class Data(Dataset):
  def __init__(self, X_train, y_train):

    self.X = torch.from_numpy(X_train).type(torch.float32)
    self.y = torch.from_numpy(y_train).type(torch.LongTensor)
    self.len = self.X.shape[0]
  
  def __getitem__(self, index):
    return self.X[index], self.y[index]
  
  def __len__(self):
    return self.len
  

Engine_data = pd.read_csv('modified.csv')
Engine_data = shuffle(Engine_data)
Features = Engine_data.iloc[:,:-1]
Labels = Engine_data.iloc[:,-1]
X_train, X_test, y_train, y_test = train_test_split(Features, Labels, test_size=0.2, random_state=42)
traindata = Data(X_train.values, y_train.values)
batch_size = 128
trainloader = DataLoader(traindata, batch_size=batch_size, shuffle=True, num_workers=0)



class Network(nn.Module):
  def __init__(self,input_dim,embedding_dim):
    super(Network, self).__init__()
    self.linear1 = nn.Linear(input_dim,256)
    self.linear2 = nn.Linear(256, 128)
    self.linear3 = nn.Linear(128, 64)
    self.linear4 = nn.Linear(64, 32)
    self.linear5 = nn.Linear(32, 64)
    self.linear6 = nn.Linear(64, 128)
    self.linear7 = nn.Linear(128, 256)
    self.linear8 = nn.Linear(256, embedding_dim)
    self.batch_norm_1 =nn.BatchNorm1d(num_features=256)
    self.batch_norm_2 =nn.BatchNorm1d(num_features=128)
    self.batch_norm_3 =nn.BatchNorm1d(num_features=64)
    self.batch_norm_4 =nn.BatchNorm1d(num_features=32)


  def forward(self, x):
    x = torch.relu(self.linear1(x))
    x = self.batch_norm_1(x)
    x = torch.relu(self.linear2(x))
    x = self.batch_norm_2(x)
    x = torch.relu(self.linear3(x))
    x = self.batch_norm_3(x)
    x = torch.relu(self.linear4(x))
    x = self.batch_norm_4(x)
    x = torch.relu(self.linear5(x))
    x = self.batch_norm_3(x)
    x = torch.relu(self.linear6(x))
    x = self.batch_norm_2(x)
    x = torch.relu(self.linear7(x))
    x = self.batch_norm_1(x)
    x = torch.sigmoid(self.linear8(x))
    return x
  

def weights_init(m):
    if isinstance(m, nn.Conv2d):
       torch.nn.init.kaiming_uniform_(m.weight.data)
       torch.nn.init.kaiming_uniform_(m.bias.data)
Embedding_network = Network(input_dim = 6,embedding_dim =512)
Embedding_network.apply(weights_init)

optimizer = torch.optim.Adam(Embedding_network.parameters(), lr=0.001)
PATH = './mymodel.pth'
epochs = 30
for epoch in range(epochs):
  running_loss = 0.0
  for data in tqdm(trainloader):
    inputs, labels = data
    optimizer.zero_grad()
    Output1 = Embedding_network(inputs[:,:6])
    Output2 = Embedding_network(inputs[:,6:12])
    mean_loss = torch.sum(torch.square(Output1-Output2),dim=1,keepdim=True)
    loss = (1-labels)*mean_loss + labels*torch.abs(1-mean_loss) 
    loss = torch.mean(loss)
    loss.backward()
    optimizer.step()
    running_loss += loss.item()
  print(f'Epoch:{epoch + 1} loss: {running_loss:.10f}')

  torch.save(Embedding_network.state_dict(), PATH)




Embedding_network = Network(input_dim = 6,embedding_dim =512)
Embedding_network.load_state_dict(torch.load(PATH))
File ="Embedding.csv"

Engine_data = pd.read_csv('modified.csv')
Features = Engine_data.iloc[:,:-1]
Labels = Engine_data.iloc[:,-1]



File ="Embedding.csv"
df = pd.read_csv('engine_data.csv')
features = df[['Engine rpm', 'Lub oil pressure', 'Fuel pressure', 'Coolant pressure', 'lub oil temp', 'Coolant temp']]
target = df[['Engine Condition']]

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.15, random_state=42)
Features = X_train
Labels =y_train



for i in range(Features.shape[0]):
    input = Features.iloc[i]
    input = np.array(input).astype(np.float32)
    input = torch.from_numpy(input).unsqueeze(dim=0)
    Embedding_network.eval()
    Embedding = Embedding_network(input)
    Embedding =Embedding.detach().cpu().numpy()
    Embedding = np.append(Embedding,Labels.iloc[i])
    Embedding= pd.DataFrame(Embedding.reshape(1,Embedding.shape[0]))
    Embedding.to_csv(File,mode='a',header=False,index=False)


Embedding = pd.read_csv('Embedding.csv')
Embedding = Embedding.values
Class_0_Embedding = Embedding[Embedding[:,-1]==0][:,:-1]
Class_1_Embedding = Embedding[Embedding[:,-1]==1][:,:-1]

Class_0_mean = np.mean(Class_0_Embedding,axis=0)
Class_1_mean = np.mean(Class_1_Embedding,axis=0)

Features = X_test
Labels =y_test
y_pred = []
for i in range(Features.shape[0]):
    input = Features.iloc[i]
    input = np.array(input).astype(np.float32)
    input = torch.from_numpy(input).unsqueeze(dim=0)
    Embedding_network.eval()
    Embedding = Embedding_network(input)
    Embedding =Embedding.detach().cpu().numpy()
    if np.sum(np.square(Embedding - Class_0_mean)) >  np.sum(np.square(Embedding - Class_1_mean)):
       y_pred.append(1)
    else:
       y_pred.append(0)

y_test_n = np.array(y_test).reshape(1,-1)[0]
y_pred = np.array(y_pred)
correct = np.count_nonzero(y_test_n == y_pred)
accuracy = correct / len(y_test_n)
print(accuracy)
```

## Stacking

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import StackingClassifier, AdaBoostClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.svm import SVC


base_learners = [
    ("ada", AdaBoostClassifier(n_estimators=200)),
    ("svm", SVC(C=10, kernel='linear', gamma=0.1)),
    ("rf", RandomForestClassifier(n_estimators=200)),
    ("xg", XGBClassifier(max_depth=10, learning_rate=0.1, n_estimators=200))  # Adjust hyperparameters as needed
]

# Define final classifier (meta-learner)
final_estimator = LogisticRegression()

# Stacking classifier with cross-validation for final_estimator
stacking_clf = StackingClassifier(estimators=base_learners, final_estimator=final_estimator, cv=3)

# Train the stacking model
stacking_clf.fit(X_train, y_train)

# Make predictions on unseen testing data
y_pred = stacking_clf.predict(X_test)

# Evaluate model performance on testing data (e.g., using accuracy score)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("StackingClassifier Accuracy:", accuracy)
```
